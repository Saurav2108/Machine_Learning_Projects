# Problem Statement 

In this case study we want to build a classifier that can predict whether a new sms is a spam or not. 

Dataset = SMSSpamcollection

Developer = Saurav Bhagat


```python
#Import required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sms = pd.read_csv ("SMSSpamCollection", usecols =[0,1],skiprows=1, names=["label","message"], sep="\t")
sms.head()
```

```python
sms['label'] =sms['label'].map(dict(spam=1, ham=0))
sms.head()
```

```python
# Data Split
```

```python
x = sms['message'].to_list()
y = sms['label'].to_list()
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(lowercase = True, stop_words = 'english')
x_train = cv.fit_transform(x)
print(x_train)
```

```python
# Modelling
```

```python
#model = MultinomialNB(class_prior = [0.6, 0.4]) #we can set prior probabilities. Otherwise, the model will calculate these from the training data.
model = MultinomialNB()
model.fit(x_train, y)
```

```python
# lets check the model on new messages
```

```python
test_msg = ['Ok Im joking', 'You won a prize!', 'Need a new mobile?','Bye Brother']
x_test = cv.transform(test_msg)  # Use transform instead of fit_transform
model.predict(x_test)
```
